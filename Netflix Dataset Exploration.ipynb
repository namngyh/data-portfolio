```python
# Import library

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Add and print data

data = pd.read_csv('Netflix Dataset.csv')
print(data.head(7))
```

```python
# For 'House of Cards', what is the Show Id and Who is the Director of this show ?

q1 = data[data['Title'] == 'House of Cards']
print("For 'House of Cards'")
print("Show Id:", q1['Show_Id'].values[0])
print("Director:", q1['Director'].values[0])
```

```python
# In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph
data['Release_Date'] = pd.to_datetime(data['Release_Date'], errors='coerce')
data['Year'] = data['Release_Date'].dt.year
year_counts = data['Year'].value_counts().sort_index()
release_counts = data['Year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
year_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of TV Shows & Movies Released')
plt.title('Number of TV Shows & Movies Released Each Year')
plt.show()

max_year = year_counts.idxmax()
max_count = year_counts.max()
print(f"The highest number of releases was in {max_year} with {max_count} releases.")
```

```python
# How many Movies & TV Shows are in the dataset? Show with Bar Graph

type_counts = data['Type'].value_counts()

plt.figure(figsize=(6,4))
type_counts.plot(kind='bar', color=['orange', 'purple'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Number of Movies & TV Shows in the Dataset')
plt.xticks([])
plt.show()

print(type_counts)
```

```python
# Show all the Movies that were released in year 2000

movies_2000 = data[(data['Type'] == 'Movie') & (data['Year'] == 2000)]
print(movies_2000)
```

```python
#Show only the Titles of all TV Shows that were released in India only

india_tv_shows = data[(data['Type'] == 'TV Show') & (data['Country'] == 'India')]
print(india_tv_shows['Title'])
```

```python
#Show Top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix

top_directors = data['Director'].value_counts().head(10)
print("Top 10 Directors with the highest number of TV Shows & Movies:")
print(top_directors)    
```

```python
#Show all the Records, where "Category is Movie and Type is Comedies" or "Country is United Kingdom"

comedy_movies_uk = data[((data['Type'] == 'Movie') & (data['Category'] == 'Comedies')) | (data['Country'] == 'United Kingdom')]
print(comedy_movies_uk)
```

```python
#In how many movies/shows, Tom Cruise was cast

tom_cruise_count = data[data['Cast'].str.contains('Tom Cruise', na=False)]#.shape[0]
print(f"Tom Cruise was cast in {tom_cruise_count} movies/shows.")   
```

```python
#What are the different Ratings defined by Netflix 

ratings = data['Rating'].unique()
print("Different Ratings defined by Netflix:")
print(ratings)
```

```python
#How many Movies got the 'TV-14' rating, in Canada

tv14_canada_count = data[(data['Rating'] == 'TV-14') & (data['Country'] == 'Canada')]#.shape[0]
print(f"Number of Movies with 'TV-14' rating in Canada: {tv14_canada_count}")
```

```python
#How many TV Shows got the 'R' rating, after year 2018

tv_shows_r_rating = data[(data['Rating'] == 'R') & (data['Type'] == 'TV Shows') & (data['Year'] > 2018)]
print(f"Number of TV Shows with 'R' rating after 2018: {tv_shows_r_rating.shape[0]}")
```

```python
#What is the maximum duration of a Movie/Show on Netflix

max_duration = data['Duration'].max()
print(f"The maximum duration of a Movie/Show on Netflix is: {max_duration} minutes")
```

```python
# Which individual country has the Highest No. of TV Shows

country_tv_show_counts = data[data['Type'] == 'TV Shows']['Country'].value_counts()
top_country = country_tv_show_counts.idxmax()
top_count = country_tv_show_counts.max()
print(f"The country with the highest number of TV Shows is {top_country} with {top_count} TV Shows.")
```

```python
#How can we sort the dataset by Year ?

sorted_data = data.sort_values(by='Year')
print("Dataset sorted by Year:")
print(sorted_data[['Title', 'Year']].head(10)) 
```

```python
#Find all the instances where: Category is 'Movie' and Type is 'Dramas' or Category is 'TV Show' & Type is 'Kids' TV'
movie_dramas_kids_tv = data[((data['Category'] == 'Movie') & (data['Type'] == 'Dramas')) | ((data['Category'] == 'TV Show') & (data['Type'] == 'Kids TV'))]
print("Instances where Category is 'Movie' and Type is 'Dramas' or Category is 'TV Show' & Type is 'Kids TV':")
print(movie_dramas_kids_tv[['Title', 'Category', 'Type']])  
```
